```python
# Importem les llibreries necessàries:
import pandas as pd
import numpy as np

# Carreguem el conjunt de dades:
data = pd.read_csv('data.csv')

# Mostrem per pantalla les primeres files:
data.head()
```

```python
# Afegim una columna numèrica que indiqui que cada files es tracta d'una trucada diferent:
data.insert(loc=0, column='Trucades', value=1)

# Mostrem el canvi per pantalla:
data.head()
```

```python
# Convertim la columna referent a la hora de les trucades de format 12h a format de 24h:
data['Hora'] = pd.to_datetime(data['Hora']).dt.strftime('%H:%M:%S')

# Seleccionem els registres per franja horària d'interès
nit = data.loc[(data['Hora'] > '00:00:00') & (data['Hora'] < '08:00:00')]
mati = data.loc[(data['Hora'] > '08:00:00') & (data['Hora'] < '14:00:00')]
tarda = data.loc[(data['Hora'] > '14:00:00') & (data['Hora'] < '20:00:00')]
vespre = data.loc[(data['Hora'] > '20:00:00') & (data['Hora'] < '23:59:00')]

# Els substituim per les hores agrupades en franjes:
nit['Hora'] = 'Nit'
mati['Hora'] = 'Matí'
tarda['Hora'] = 'Tarda'
vespre['Hora'] = 'Vespre'

# Els agrupem de nou per treballar amb el dataset complet:
data = pd.concat([nit, mati, tarda, vespre])

# Mostrem el canvi per pantalla:
data.head()
```

```python
# Obtenció del nombre total de trucades per any tot diferenciant les que han sigut per motiu de violència masclista.

# Seleccionem les columnes a utilitzar:
trucades_anuals = data[['Trucades', 'Any', 'Motiu Trucada']]

# Convertim el motiu de les trucades a un camp numèric en què el número 1 indica les referents a violència masclista:
trucades_anuals.loc[trucades_anuals['Motiu Trucada'] == 'Violència masclista', 'Motiu Trucada'] = 1
trucades_anuals.loc[trucades_anuals['Motiu Trucada'] == 'No es tracta de violència', 'Motiu Trucada'] = 0
trucades_anuals['Motiu Trucada'] = trucades_anuals['Motiu Trucada'].fillna(0)

# Canviem el nom de la columna per no equivocar-nos a l'hora d'interpretar els resultats:
trucades_anuals = trucades_anuals.rename(columns={'Motiu Trucada': 'Trucades per violència masclista'})

# Les agrupem en funció de l'any de la trucada:
trucades_anuals = trucades_anuals.groupby(['Any']).sum()

# Mostrem per pantalla el resultat:
trucades_anuals
```

```python
# Obtenció del nombre de trucades mensuals durant l'any 2020.

# Seleccionem les columnes a utilitzar:
trucades_mensuals_2020 = data[['Trucades', 'Any', 'Mes']]

# Seleccionem únicament els registres referents a l'any 2020:
trucades_mensuals_2020 = trucades_mensuals_2020.loc[trucades_mensuals_2020['Any'] == 2020]

# Les agrupem en funció del mes:
trucades_mensuals_2020 = trucades_mensuals_2020.groupby(['Mes']).sum()

# Les mostrem per pantalla:
trucades_mensuals_2020.loc[:, trucades_mensuals_2020.columns != 'Any']
```

```python
# Obtenció de les trucades agrupades en funció del dia de la setmana i l'any.

# Seleccionem les columnes a utilitzar:
trucades_setmanals = data[['Trucades', 'Any', 'Día setmana']]

# Les agrupem en funció del dia de la setmana i l'any:
trucades_setmanals = trucades_setmanals.groupby(['Día setmana', 'Any']).sum()

# Les mostrem per pantalla:
trucades_setmanals
```
